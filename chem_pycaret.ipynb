```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
from pycaret.classification import ClassificationExperiment, setup, compare_models, tune_model, create_model

CSV_PATH = "https://docs.google.com/spreadsheets/d/e/2PACX-1vROupcEcGFWafl16RmdNcSg7J3ZfCyD1socrrhGBwE0JBD_G7GN7r8YvYKSvyQzsxRW19MYpLkRClrU/pub?gid=0&single=true&output=csv"

df = pd.read_csv(CSV_PATH)

# Setup the PyCaret classification experiment
exp = ClassificationExperiment()

# Setup the dataset and target variable
s = setup(df, target='y', session_id=123)

# Compare baseline models and select the best
best = compare_models()

# Print out the results
print(best)

# Tune the KNN model
knn = create_model('knn')

# Define the hyperparameter grid for tuning
params = {
    'n_neighbors': np.arange(1, 51),
    'weights': ['uniform', 'distance']
}

# Tune the KNN model using the defined grid and F1 metric
tuned_knn = tune_model(knn, optimize='F1', fold=5, n_iter=10, early_stopping=True, early_stopping_max_iters=10, custom_grid=params)

# Print the tuned KNN model's parameters
print(tuned_knn)
```
